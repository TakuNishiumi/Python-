# matplotlibでDS9のように星を見えるようにする

通常matplotlibなどで天体の画像を表示するだけだと、実際には複数の天体が写っていたとしても、明るい数天体しか見えないことが多い

以下がその例

```python
%matplotlib inline
from matplotlib import pyplot as plt 
from astropy.visualization import ZScaleInterval
import astropy as ap
import astropy.io.fits 
filename = "test.fits"
data = ap.io.fits.getdata(filename)
plt.imshow(data)
```

## ds9のzscaleに関係した表示(計算式はds9と同じではないが似た計算方法)

以下の方法を用いてzscale表示すれば暗い星までしっかりと表示させることができる。

ただし、この方法は全て値が0-1までに規格化されてしまうので、注意が必要です。多分、元の値を保持しておく方法は存在していると思うのですが、自分はまだ理解していません。ご存知であればコメントをください。

```python
%matplotlib inline
zscale = ZScaleInterval()
data = zscale(data)
plt.imshow(data)
```
